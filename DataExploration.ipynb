```python
from sklearn.decomposition import PCA

# Prepare the data by removing 'fetal health'
X = data_frame.drop(columns='fetal_health')

# Standardize the data
from sklearn.preprocessing import StandardScaler
X_scaled = StandardScaler().fit_transform(X)

# Perform PCA
pca = PCA()
pca.fit(X_scaled)

# Calculate the cumulative explained variance
cumulative_variance = pca.explained_variance_ratio_.cumsum()

# Line Plot the cumulative explained variance
plt.figure(figsize=(10, 6))
plt.plot(range(1, len(cumulative_variance) + 1), cumulative_variance, marker='o', linestyle='--')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance')
plt.title('Cumulative Explained Variance by Number of Principal Components')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python

#Find the number of PCA components required to explain 95% of the variance
min_components_95 = (cumulative_variance >= 0.95).argmax() + 1

print(f"Minimum number of components to explain 95% variance: {min_components_95}")

# Transform the data using the chosen number of components
pca = PCA(n_components=min_components_95)
X_pca = pca.fit_transform(X_scaled)

# Create a DataFrame for the PCA-transformed data, including the target
pca_df = pd.DataFrame(X_pca, columns=[f'PC{i+1}' for i in range(min_components_95)])
pca_df['fetal_health'] = data_frame['fetal_health'].values

# Display the resulting DataFrame
print(pca_df.head())

# Convert to a tensor (for training a model later)
import torch
pca_tensor = torch.tensor(pca_df.values, dtype=torch.float32)

print("Tensor shape:", pca_tensor.shape)
```

```python
import torch
import torch.nn as nn
import torch.optim as optim

# Define the model
class FetalHealthModel(nn.Module):
    def __init__(self):
        super(FetalHealthModel, self).__init__()
        self.fc1 = nn.Linear(14, 14)  # First fully connected layer 14 -> 14
        self.relu = nn.ReLU()         # Activation function
        self.fc2 = nn.Linear(14, 3)   # Output layer 14 -> 3 
        self.dropout = nn.Dropout(0.2)
    def forward(self, x):
        x = self.fc1(x)
        x = self.relu(x)
        x = self.dropout(x)
        x = self.fc2(x)
        return x
```

```python

def train_and_evaluate(model, criterion, train_loader, test_loader, num_epochs=10, lr=0.005):
    """
    Trains and evaluates the model.
    
    Parameters:
    - model: the PyTorch model to train
    - criterion: the loss function to use
    - train_loader: DataLoader for the training dataset
    - test_loader: DataLoader for the testing dataset
    - num_epochs: number of epochs to train
    - lr: learning rate for the optimizer
    """
    # Define the optimizer
    optimizer = optim.Adam(model.parameters(), lr=lr)

    for epoch in range(num_epochs):
        model.train()  # Set the model to training mode
        for batch_features, batch_labels in train_loader:
            optimizer.zero_grad()  # Zero the parameter gradients

            outputs = model(batch_features)  # Forward pass
            loss = criterion(outputs, batch_labels)  # Compute the loss

            loss.backward()  # Backward pass
            optimizer.step()  # Optimization step

        # Evaluate on test data after each epoch
        model.eval()  # Set the model to evaluation mode
        correct = 0
        total = 0
        false_negatives = 0  # Counter for false negatives
        with torch.no_grad():  # No need to compute gradients during evaluation
            for batch_features, batch_labels in test_loader:
                outputs = model(batch_features)
                _, predicted = torch.max(outputs, 1)  # Get the index of the max output
                total += batch_labels.size(0)
                correct += (predicted == batch_labels).sum().item()

                # Count false negatives (predicted 1 but actually 2 or 3)
                false_negatives += ((predicted == 0) & ((batch_labels == 1) | (batch_labels == 2))).sum().item()

        accuracy = 100 * correct / total
        false_negative_percentage = 100 * false_negatives / total
        print(f'Epoch [{epoch+1}/{num_epochs}], Loss: {loss.item():.4f}, Test Accuracy: {accuracy:.2f}%, False Negatives: {false_negative_percentage:.2f}%')

    print("Training complete!")
```

```python
from sklearn.model_selection import train_test_split
from torch.utils.data import TensorDataset, DataLoader

# Split the PCA DataFrame into features and labels
features = pca_df.drop(columns=['fetal_health']).values
labels = pca_df['fetal_health'].values

# Convert to tensors
features_tensor = torch.tensor(features, dtype=torch.float32)
labels_tensor = torch.tensor(labels-1, dtype=torch.long)
# Split the data into training and test sets 90/10
train_features, test_features, train_labels, test_labels = train_test_split(
    features_tensor, labels_tensor, test_size=0.1, random_state=42
)

# Create DataLoaders for training and testing
train_dataset = TensorDataset(train_features, train_labels)
test_dataset = TensorDataset(test_features, test_labels)

train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=32, shuffle=False)


# Instantiate the model, define the loss function and optimizer
model = FetalHealthModel()
criterion = nn.CrossEntropyLoss()  # Loss for predicting classes (no weights)
optimizer = optim.Adam(model.parameters(), lr=0.005)

train_and_evaluate(model,criterion, train_loader, test_loader, num_epochs=10, lr=0.005)
```

```python
import numpy as np
# Count the number of occurrences of each label
label_counts = np.bincount(labels_tensor.numpy())

# Calculate the percentage (frequency) of each label
total_labels = len(labels_tensor)
label_percentages = label_counts / total_labels

# Convert percentages to weights (inverse of the percentage)
weights = 1.0 / label_percentages

# Normalize weights to avoid extremely large values
weights = weights / weights.sum()

# Convert weights to a tensor
weights_tensor = torch.tensor(weights, dtype=torch.float32)

# Instantiate the custom criterion with a specified penalty weight
criterion = nn.CrossEntropyLoss(weight=weights_tensor)

model = FetalHealthModel()
# Train the model using the cross entropy with weights to hopefully reduce false negatives
train_and_evaluate(model, criterion, train_loader, test_loader, num_epochs=7,lr=0.001)
```
